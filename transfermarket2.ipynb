```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time


headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36'}

url = "https://www.transfermarkt.com/premier-league/startseite/wettbewerb/GB1"

r = requests.get(url, headers=headers)
r.status_code

```

```python
soup = bs(r.content, 'html.parser')
data1 = soup.find('div', id='yw1')
data2 = data1.find_all('tr', class_=['odd', 'even'])

clubs = []
squad = []
mv = []
totalMv = []

```

```python
for data in data2:
    info = data.find_all("td")
    
    clubs.append(info[1].get_text())
    squad.append(info[3].get_text())
    mv.append(info[6].get_text())
    totalMv.append(info[7].get_text())

# print(clubs)
# print(squad)

```

```python
df = pd.DataFrame(
    {
        "Clubs":clubs,
        "Squad":squad,
        "MV":mv,
        "Total Mv":totalMv
    }
)

df
```
